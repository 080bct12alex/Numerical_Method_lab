```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 4*x - 9

def df(x):
    return 3*x**2 - 4

def newtonRaphson(x0, toleration=0.0001, iteration=0,  max_iterations=10): #recursive as it uses recursion (calls itself until convergence or max iterations).
    if df(x0) == 0:
        print(f"Derivative zero at x = {x0}. Cannot proceed.")
        return None

    x1 = x0 - f(x0) / df(x0)
    iteration += 1

    if abs(f(x1)) <= toleration or abs(x1 - x0) < toleration:    
        print(f"Root found: {x1} after {iteration} iterations.")
        return x1

    if iteration >= max_iterations:
        print(f"Max iterations reached. Approximate(last) root: {x0}")
        return x0    #can use either x0 or x1 at the end, but which one makes more sense depends on how your loop updates the variables.as x1 is the new approximation calculated in the current iteration.x0 is updated to x1 at the end of the loop for the next iteration.After the last iteration, x0 holds the latest approximation because you assigned x0 = x1 before looping again. So:If you return after the loop ends, x0 contains the last approximation.x1 also holds the last computed value, but x0 is typically better because that’s what you carry forward.
                                                      #Summary: You can use either if you’re careful, but using x0 is cleaner because:
                                                                        #It’s the variable that consistently holds the current guess.
                                                                        #It matches how the algorithm progresses.
                     # Return x1 would also work because x1 was just calculated — but you don’t use x1 after the loop ends, so it's slightly less consistent.

    return newtonRaphson(x1, toleration, iteration, max_iterations)


root = newtonRaphson(2)
print("Root:", root)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 4*x - 9

def df(x):
    return 3*x**2 - 4

def newtonRaphson(x0, toleration=0.0001 , iteration = 0, max_iterations=10):    #iterative
   
    for i in range(max_iterations):
        if df(x0) == 0:
            print(f"Derivative zero at x = {x0}. Cannot proceed.")
            return None

        x1 = x0 - f(x0) / df(x0)
        iteration += 1

        if abs(f(x1)) <= toleration or abs(x1 - x0) < toleration:        #Criterion	for stopping 
                                                                         #abs(f(x1)) <= tol	    Root found (function zero)	
                                                                         # abs(x1 - x0) < tol	Estimates converged	(if the estimate isn’t changing significantly.)



            print(f"Root found: {x1} after {iteration} iterations.")
            return x1

        x0 = x1

    print(f"Max iterations reached. Last approximation: {x0}")
    return x0

    


root = newtonRaphson(2)
print("Root:", root)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- ODE Function ---
def dydx(x, y):
    return y - x**2 + 1

# --- Euler Method ---
def euler(f, x0, y0, h, n):
    x_vals = [x0]
    y_vals = [y0]
    for i in range(n):     # _ in for _ in range(n):  _ is a convention in Python to say: “I’m ignoring the loop variable.” or can use  for i in range(n):   
        y0 += h * f(x0, y0)
        x0 += h
        x_vals.append(x0)
        y_vals.append(y0)
        print(f"Euler Step {i+1}: x = {x0:.4f}, y = {y0:.4f}")
    return x_vals, y_vals

# --- Runge-Kutta 4th Order Method ---
def rk4(f, x0, y0, h, n):
    x_vals = [x0]
    y_vals = [y0]
    for i in range(n):
        k1 = h * f(x0, y0)
        k2 = h * f(x0 + h/2, y0 + k1/2)
        k3 = h * f(x0 + h/2, y0 + k2/2)
        k4 = h * f(x0 + h, y0 + k3)
        y0 += (k1 + 2*k2 + 2*k3 + k4) / 6
        x0 += h
        x_vals.append(x0)
        y_vals.append(y0)
        print(f"RK4 Step {i+1}: x = {x0:.4f}, y = {y0:.4f}")
    return x_vals, y_vals

# --- Exact Solution ---
def exact(x):
    return (x + 1)**2 - 0.5 * np.exp(x)    #This is a first-order linear ODE, and we solve it using the integrating factor (I.F.) method

# --- Parameters ---
x0 = 0
y0 = 0.5
h = 0.5
xn = 2
n = int((xn - x0) / h)

# --- Solve ---
x_euler, y_euler = euler(dydx, x0, y0, h, n)
x_rk4, y_rk4 = rk4(dydx, x0, y0, h, n)
x_exact = np.linspace(x0, xn, 100)
y_exact = exact(x_exact)

# --- Plot ---
plt.plot(x_euler, y_euler, 'o--', label='Euler Method')
plt.plot(x_rk4, y_rk4, 's-', label='RK4 Method')
plt.plot(x_exact, y_exact, 'k-', label='Exact Solution')

# --- Mark Final Values --- ( The [-1] syntax in Python means “the last element of the list (or array).)
plt.plot(x_euler[-1], y_euler[-1], 'ro', label=f'Euler Final: y = {y_euler[-1]:.4f}')
plt.plot(x_rk4[-1], y_rk4[-1], 'go', label=f'RK4 Final: y = {y_rk4[-1]:.4f}')
plt.plot(xn, exact(xn), 'ko', label=f'Exact Final: y = {exact(xn):.4f}')

plt.title("ODE Solution: dy/dx = y - x^2 + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def lagrange_interpolation(x_points, y_points, x):
    n = len(x_points)
    result = 0.0

    for i in range(n):
        term = y_points[i]
        for j in range(n):
            if j != i:
                term *= (x - x_points[j]) / (x_points[i] - x_points[j])
        result += term

    return result


# Given data
x_points = [0, 1, 2, 4]
y_points = [0, 1, 4, 16]
x = 3

# Interpolated value at x
y_interp = lagrange_interpolation(x_points, y_points, x)
print("Interpolated value at x =", x, "is", y_interp)

# Generate interpolated curve 
x_vals = np.linspace(min(x_points), max(x_points), 100)
y_vals = [lagrange_interpolation(x_points, y_points, xi) for xi in x_vals]

# Plot only interpolated curve and point
plt.plot(x_vals, y_vals, label='Interpolated Curve', color='blue')
plt.plot(x, y_interp, 'go', label=f'Interpolated Point ({x}, {y_interp:.2f})')

plt.title('Lagrange Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
