```python
import numpy as np
import matplotlib.pyplot as plt



x = np.array([2, 4, 6, 8, 10, 12])
y = np.array([3, 5, 10, 18, 48, 90])


n = len(x)
sum_x = np.sum(x)
sum_x2 = np.sum(x**2)
sum_xy = np.sum(x * y)
sum_y = np.sum(y)
m = (n*sum_xy - sum_x*sum_y)/(n*sum_x2 - (sum_x)**2)
c = (sum_y - m*sum_x)/n


print(f"Equation is y = {m:.4f}x + {c:.4f}.")

y_pred = m * x + c

# Plotting the actual data points
plt.scatter(x, y, color='red', label='Actual data')

# Plotting the regression line
plt.plot(x, y_pred, color='blue', label='Fitted line')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Least Squares Fit')
plt.legend()
plt.grid(True)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt


x = np.array([2, 4, 6, 8, 10, 12])
y = np.array([3, 5, 10, 18, 48, 90])


Y = np.log(y)


n = len(x)
sum_x = np.sum(x)
sum_x2 = np.sum(x**2)
sum_Y = np.sum(Y)
sum_xY = np.sum(x * Y)


b = (n * sum_xY - sum_x * sum_Y) / (n * sum_x2 - sum_x**2)
ln_A = (sum_Y - b * sum_x) / n
A = np.exp(ln_A)

print(f"Equation is: y = {A:.6f} * e^({b:.6f} * x)")


y_pred = A * np.exp(b * x)

# Plotting
plt.scatter(x, y, color='red', label='Original data')
plt.plot(x, y_pred, color='blue', label='Exponential Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Least Squares Fit')
plt.legend()
plt.grid(True)
plt.show()
```
